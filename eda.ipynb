```python
text_corpus = community['question'].astype(str) + " " + community['answer'].astype(str) + " " + community['wrong_answer'].astype(str) + community['title'].astype(str) 

# Convert to a single string
full_text = " ".join(text_corpus)
print(len(full_text))
```

```python
from nltk.corpus import stopwords
 
nltk.download('stopwords')
print(stopwords.words('english'))
```

```python
import re
from nltk.tokenize import word_tokenize

tokens = word_tokenize(full_text.lower())
#tokens = [re.sub(r'\W+', '', token) for token in tokens if token.isalpha()]  # Remove non-alphabetic tokens
tokens = [re.sub('W+', '', token) for token in tokens if token.isalpha() and token not in stopwords.words('english')]
```

```python
from collections import Counter

# Count token frequencies
token_counts = Counter(tokens)

# Display the most common tokens
print(token_counts.most_common(10))
```

```python
vocab = {token for token, count in token_counts.items() if count >= 2}

# Add special tokens
vocab_dict = {token: idx for idx, token in enumerate(vocab, start=2)}
vocab_dict['<PAD>'] = 0  # Padding token
vocab_dict['<UNK>'] = 1  # Unknown token

print(f"Vocabulary size: {len(vocab_dict)}")
```
